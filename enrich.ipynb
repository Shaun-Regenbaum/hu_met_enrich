```python
# Initialize a GProfiler instance
gp = GProfiler()
```

```python
import requests

def get_gene_ids(pathway_id):
    pathway_url = "http://rest.kegg.jp/link/ko/{}".format(pathway_id)
    response = requests.get(pathway_url)
    gene_ids = []
    if response.status_code == 200:
        for line in response.text.strip().split("\n"):
            gene_id = line.split("\t")[1].split(":")[1]
            gene_ids.append(gene_id)
    return gene_ids

gene_ids = get_gene_ids("map01100")
print(gene_ids)
```

```python

# Perform pathway enrichment analysis on the gene IDs for KEGG metabolic pathways
# domain_scope=["path:map01100"]


result = gp.profile(query=gene_ids, organism='hsapiens', sources=['KEGG'] )
```
